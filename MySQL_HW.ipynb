```python
import pymysql
import prettytable
import IPython.display as dp
import sys

pt = prettytable.PrettyTable(["編號", "姓名", "生日", "住址"], encoding = "UTF-8")
pt.align["編號"]="l"
pt.align["姓名"]="l"
pt.align["生日"]="l"
pt.align["住址"]="l"

h = input("請輸入主機位置:")
u = input("請輸入資料庫帳號：")
p = input("請輸入資料庫密碼：")
po = int(input("請輸入資料庫的Port："))
b = input("請輸入資料庫的名稱：")

# c = pymysql.connect(
#   host = "0.tcp.jp.ngrok.io",
#   user = "root",
#   passwd = "",
#   db = "HW",
#   charset = "utf8",
#   port = 15000
# )

com = 5
while not com == 0:
  while not 0 < com <= 4:
    print("(0) 離開程式")
    print("(1) 顯示會員列表")
    print("(2) 新增會員資料")
    print("(3) 更新會員資料")
    print("(4) 刪除會員資料")
    com = int(input("指令："))
    dp.clear_output(False)


  # (1) 顯示會員列表
  while com == 1:
      c1 = pymysql.connect(
      host = h,
      user = u,
      passwd = p,
      db = b,
      charset = "utf8",
      port = po
      )

      p1 = c1.cursor()
      p1.execute("SELECT * FROM `member`")
      pt.clear_rows()
      for d1 in p1.fetchall():
        pt.add_row([d1[0], d1[1], d1[2], d1[3]])
      c1.close()

      print(pt)
      print("(0) 離開程式")
      print("(1) 顯示會員列表")
      print("(2) 新增會員資料")
      print("(3) 更新會員資料")
      print("(4) 刪除會員資料")
      com = int(input("指令："))
      dp.clear_output(False)

  # (2) 新增會員資料
  while com == 2:
      c2 = pymysql.connect(
      host = h,
      user = u,
      passwd = p,
      db = b,
      charset = "utf8",
      port = po
      )

      p2 = c2.cursor()
      p2.execute("INSERT INTO `member`(`name`, `birthday`, `address`) VALUE(%(a)s, %(b)s, %(c)s)",
              {
                "a":input("請輸入會員姓名:"),
                "b":input("請輸入會員生日:"),
                "c":input("請輸入會員住址:")
              }
          )
      c2.commit()
      c2.close()

      dp.clear_output(True)
      print("(0) 離開程式")
      print("(1) 顯示會員列表")
      print("(2) 新增會員資料")
      print("(3) 更新會員資料")
      print("(4) 刪除會員資料")
      com = int(input("指令："))
      dp.clear_output(False)

  # (3) 更新會員資料
  while com == 3:
      c3 = pymysql.connect(
      host = h,
      user = u,
      passwd = p,
      db = b,
      charset = "utf8",
      port = po
      )

      p3 = c3.cursor()
      p3.execute("SELECT * FROM `member`")
      pt.clear_rows()
      for d3 in p3.fetchall():
        pt.add_row([d3[0], d3[1], d3[2], d3[3]])
      print(pt)

      p3.execute("UPDATE `member` SET `name` = %(b)s, `birthday` = %(c)s, `address` = %(d)s WHERE `id` = %(a)s",
                            {
                "a":int(input("請選擇你要修改的資料編號:")),
                "b":input("請輸入會員姓名:"),
                "c":input("請輸入會員生日:"),
                "d":input("請輸入會員住址:")
              }
              )
      c3.commit()
      c3.close()

      dp.clear_output(True)
      print("(0) 離開程式")
      print("(1) 顯示會員列表")
      print("(2) 新增會員資料")
      print("(3) 更新會員資料")
      print("(4) 刪除會員資料")
      com = int(input("指令："))
      dp.clear_output(False)

  # (4) 刪除會員資料
  while com == 4:
      c4 = pymysql.connect(
      host = h,
      user = u,
      passwd = p,
      db = b,
      charset = "utf8",
      port = po
      )

      p4 = c4.cursor()
      p4.execute("SELECT * FROM `member`")
      pt.clear_rows()
      for d4 in p4.fetchall():
        pt.add_row([d4[0], d4[1], d4[2], d4[3]])
      print(pt)

      p4.execute("DELETE FROM `member` WHERE `id` = %(a)s",
                            {
                "a":int(input("請選擇你要刪除的資料編號:")),
              }
              )
      c4.commit()
      c4.close()

      dp.clear_output(True)
      print("(0) 離開程式")
      print("(1) 顯示會員列表")
      print("(2) 新增會員資料")
      print("(3) 更新會員資料")
      print("(4) 刪除會員資料")
      com = int(input("指令："))
      dp.clear_output(False)

# (0) 離開程式
if com == 0:
  sys.exit()
```
